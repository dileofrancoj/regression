```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Objetivo:
Crear una recta de regresión a partir de una nube de puntos. Esta recta es una en particular correspondiente a una familia de rectas que minimiza el error cuadrático medio

```python
X = np.array([45000,3000,3100,30000,14000,50000,16000,70000])
y_house_price = np.array([65000,10000,9500,15000,30000,100000,28000,150000])
```

```python
plt.scatter(X,y_house_price)
plt.show()
```

```python
house_model = LinearRegression(normalize=True)
```

```python
x = np.expand_dims(X,1) # [[45000], .... ] .fit necesita 2 dimensiones para poder calcular los parámetros
model = house_model.fit(x,y_house_price)
```

```python
m = model.coef_ # Slope
b = model.intercept_ #Y intercept
```

```python
regression_line_values = m*X + b
plt.scatter(X,y_house_price)
plt.xlabel('Superficie')
plt.ylabel('Precio')
plt.plot(X,regression_line_values,'r')
```

```python
def predictValue(val):
    return ((np.transpose(val) * m) + b)
```
