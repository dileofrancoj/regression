```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
celsius = np.array([-40,-10,0,8,15,22,38], dtype=float)
fahrenheit = np.array([-40,14,32,46,59,72,100], dtype=float)
```

```python
X = celsius.reshape(-1,1)
Y = fahrenheit.reshape(-1,1)

lr = LinearRegression()
model = lr.fit(X,Y)
```

```python
m = model.coef_
b = model.intercept_
y = m*X + b

plt.xlabel("Celsius")  
plt.ylabel("Fahrenheit")  
plt.title("Celsius to Fahrenheit")  
plt.plot(celsius,fahrenheit, color="red", label="Origina data", linewidth=1)
plt.plot(celsius,y, color="blue", label="regression line", linewidth=1)

plt.show()
```

```python
mse = np.sqrt(mean_squared_error(y,fahrenheit))
r2 = r2_score(y,fahrenheit)
print("mse: ", mse)
print("r2: ", r2)
```
